### Code SQL de la vue et de la procÃ©dure stockÃ©e

```python
#ProcÃ©dure

CREATE PROCEDURE nb_hashtags_hour (IN hashtag VARCHAR(50)) BEGIN SELECT tweets.day, tweets.hour, COUNT(tweets.hashtags) AS 'Nb_hashtags' FROM tweets WHERE tweets.hashtags = hashtag GROUP BY tweets.day, tweets.hour ORDER BY tweets.day; END ;

CALL nb_hashtags_hour('guerre');
```

```python
#View

CREATE VIEW v_tweet AS SELECT tweets.hashtags, tweets.retweetcount, users.username FROM tweets JOIN users ON tweets.userid = users.userid;

SELECT * FROM v_tweet;
```
