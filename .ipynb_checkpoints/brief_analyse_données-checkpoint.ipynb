```python
df_fr = pd.read_csv("tweet_fr_ukraine.csv")
df_fr.head()
```

```python
df_fr.shape
```

```python
df_user = df_fr.loc[:, ["userid", "username", "following", "followers"] ]
df_user.reset_index()
```

```python
df_tweet = df_fr.loc[:, ["text", "hashtags", "retweetcount", "extractedts", "userid"] ]
df_tweet
```

```python
df_tweet['hashtags']=df_tweet['hashtags'].str.split(expand=True)[1]
df_tweet.head()
```

```python
df_tweet["extractedts"]= pd.to_datetime(df_tweet["extractedts"])
```

```python
df_tweet.info()
```

```python
#Ann√©e
df_tweet['year'] = df_tweet["extractedts"].dt.year

#Mois
df_tweet['month'] = df_tweet['extractedts'].dt.month

#Jour
df_tweet['day'] = df_tweet["extractedts"].dt.day

#heure
df_tweet['hour'] = df_tweet['extractedts'].dt.hour

#minute
df_tweet["minute"] = df_tweet["extractedts"].dt.minute
```

```python
df_tweet.head()
```

```python
df_tweet.columns
```

```python
df_tweet = df_tweet[["text", "hashtags", "retweetcount", "year", "month", "day", "hour", "minute", "userid", "extractedts"]]
df_tweet.drop("extractedts", axis = 1, inplace = True)
df_tweet.head()
```

```python
df_tweet.columns
```
